```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle('UC02_df_uc_food_att.p')
```

```python
data.columns
```

```python
X = data.iloc[:, 16:]
```

```python
len(X)
```

```python
X.isnull().sum()
```

```python
X = pd.get_dummies(X, dummy_na=False, drop_first=True)
```

```python
X
```

```python
from sklearn.preprocessing import label_binarize
```

```python
y = data['stars']
def good_bad(x):
    if x >= 4.0:
        return 1
    else:
        return 0
y = y.apply(good_bad)
```

```python
y = data['stars'].astype(str)
```

```python
list(y.unique())
```

```python
y = label_binarize(y, classes=list(y.unique()))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=1000, random_state=1)
rf = rf.fit(X_train, y_train)
y_predict = rf.predict(X_test)
print roc_auc_score(y_test, y_predict, average=None)
```
